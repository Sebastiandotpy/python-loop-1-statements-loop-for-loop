```python
def multiplication_table():
    """
    Task 1: Generate the multiplication table for a given number.
    """
    number = get_positive_number("Task 1: Input a number: ")
    print(f"Task: 1 Results of multiplication by {number}:")
    for multiplier in range(1, 11):
        print(f"{number} x {multiplier} = {number * multiplier}")
    print()

def pattern_with_loop():
    """
    Task 2: Generate a pattern using a loop.
    """
    print("Task 2: Pattern with Loop:")
    for row in range(1, 10):
        print(str(row) * row)
    print()

def reverse_word():
    """
    Task 3: Reverse a word.
    """
    word = input("Task 3: Input a word to reverse: ")
    reversed_word = word[::-1]  # Reverse the word using slicing
    print("Task 3 Reversed word:", reversed_word)
    print()

def modify_sentence():
    """
    Task 4: Modify a sentence by replacing specified characters.
    """
    sentence = input("Task 4: Enter a sentence: ")
    if sentence:
        modified_sentence = sentence.replace('o', 'O')
        print("Modified sentence:", modified_sentence)
        print()
    else:
        print("Empty sentence. Please enter a valid sentence.")

def count_digits_and_letters():
    """
    Task 5: Count the number of digits and letters in a string.
    """
    characters = input("Task 5: Input your characters: ")
    digit_count = sum(char.isdigit() for char in characters)
    letter_count = sum(char.isalpha() for char in characters)
    print("Number of digits:", digit_count)
    print("Number of letters:", letter_count)
    print()

def get_positive_number(prompt):
    """
    Helper function to get a positive number from user input.
    """
    while True:
        try:
            number = int(input(prompt))
            if number <= 0:
                print("Please enter a positive number.")
            else:
                return number
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# User input for tasks
task_count = 5  # Number of tasks
completed_tasks = set()  # Track completed tasks

while len(completed_tasks) < task_count:
    choice = input("Enter task number (1-5) to see the result (or 'exit' to exit): ")

    if choice == '1' and 1 not in completed_tasks:
        multiplication_table()
        completed_tasks.add(1)
    elif choice == '2' and 2 not in completed_tasks:
        pattern_with_loop()
        completed_tasks.add(2)
    elif choice == '3' and 3 not in completed_tasks:
        reverse_word()
        completed_tasks.add(3)
    elif choice == '4' and 4 not in completed_tasks:
        modify_sentence()
        completed_tasks.add(4)
    elif choice == '5' and 5 not in completed_tasks:
        count_digits_and_letters()
        completed_tasks.add(5)
    elif choice.lower() == 'exit':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please try again.")

if len(completed_tasks) == task_count:
    print("All tasks completed. Exiting...")

```
